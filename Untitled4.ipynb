```python
#taller 8

agenda = []

while True:
    print("\n--- AGENDA DE CONTACTOS ---")
    print("1. Ver contactos")
    print("2. Agregar contacto")
    print("3. Eliminar contacto")
    print("4. Salir")

    opcion = input("Elige una opci√≥n: ")

    if opcion == "1":
        if len(agenda) == 0:
            print("No hay contactos.")
        else:
            for i in range(len(agenda)):
                print(i + 1, "-", agenda[i])
    elif opcion == "2":
        nombre = input("Nombre del contacto: ")
        telefono = input("Tel√©fono: ")
        contacto = nombre + " - " + telefono
        agenda.append(contacto)
        print("Contacto agregado.")
    elif opcion == "3":
        for i in range(len(agenda)):
            print(i + 1, "-", agenda[i])
        num = int(input("N√∫mero del contacto a eliminar: "))
        if 0 < num <= len(agenda):
            agenda.pop(num - 1)
            print("Contacto eliminado.")
        else:
            print("N√∫mero inv√°lido.")
    elif opcion == "4":
        print("Saliendo de la agenda...")
        break
    else:
        print("Opci√≥n no v√°lida.")#taller 8
```

```python
#calculadora de estadostica
def promedio(lista):
    if len(lista) == 0:
        return 0
    return sum(lista) / len(lista)

def maximo_minimo(lista):
    if len(lista) == 0:
        return None, None
    return max(lista), min(lista)

def contar(lista):
    return len(lista)

def estadisticas(lista):
    if len(lista) == 0:
        return "Lista vac√≠a"
    prom = promedio(lista)
    maxi, mini = maximo_minimo(lista)
    total = contar(lista)
    return f"Promedio: {prom}, M√°ximo: {maxi}, M√≠nimo: {mini}, Total: {total}"

def menu():
    numeros = []
    while True:
        print("\n--- Calculadora de Estad√≠sticas ---")
        print("1. Agregar n√∫mero")
        print("2. Ver promedio")
        print("3. Ver m√°ximo y m√≠nimo")
        print("4. Ver cu√°ntos hay")
        print("5. Ver estad√≠sticas completas")
        print("6. Salir")

        opcion = input("Elige una opci√≥n: ")

        if opcion == "1":
            n = float(input("Escribe un n√∫mero: "))
            numeros.append(n)
        elif opcion == "2":
            print("Promedio:", promedio(numeros))
        elif opcion == "3":
            maxi, mini = maximo_minimo(numeros)
            print("M√°ximo:", maxi, "M√≠nimo:", mini)
        elif opcion == "4":
            print("Cantidad de n√∫meros:", contar(numeros))
        elif opcion == "5":
            print(estadisticas(numeros))
        elif opcion == "6":
            print("Adi√≥s :)")
            break
        else:
            print("Opci√≥n no v√°lida")

menu()
```

```python
#liberia de funciones
def potencia(base, exponente):
    return base ** exponente

def es_primo(numero):
    if numero < 2:
        return False
    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False
    return True

def factorial(n):
    if n < 0:
        return None
    r = 1
    for i in range(1, n+1):
        r *= i
    return r

def convertir_temperatura(grados, escala):
    if escala == "C":
        return (grados * 9/5) + 32
    if escala == "F":
        return (grados - 32) * 5/9

def suma_digitos(numero):
    return sum(int(d) for d in str(abs(numero)))

print(potencia(2, 3))
print(es_primo(7))
print(factorial(5))
print(convertir_temperatura(100, "C"))
print(suma_digitos(1234))
```

```python
#dataset de deportistas
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "deporte": ["F√∫tbol", "F1", "Nataci√≥n", "Voleibol", "Basket", "Tenis"],
    "altura": [165, 182, 180, 192, 190, 195],
    "peso": [60, 72, 70, 82, 85, 85],
    "genero": ["M", "M", "M", "F", "M", "F"]
}
df = pd.DataFrame(data)
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(data=df, x="deporte", y="altura", ax=axes[0], palette="Set2")
axes[0].set_title("Altura por deporte")
sns.violinplot(data=df, x="deporte", y="peso", hue="genero",
               split=True, ax=axes[1], palette="Set1")
axes[1].set_title("Peso por g√©nero y deporte")
corr = df[["altura", "peso"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[2])
axes[2].set_title("Correlaciones f√≠sicas")
plt.tight_layout()
plt.show()
```

```python
#sistema de estudiantes

estudiantes = [
    {"id": 1, "nombre": "Ana", "grado": "10", "materias": ["Matem√°ticas", "Arte"]},
    {"id": 2, "nombre": "Luis", "grado": "9", "materias": ["Historia", "Ingl√©s"]},
    {"id": 3, "nombre": "Sof√≠a", "grado": "11", "materias": ["Biolog√≠a"]},
    {"id": 4, "nombre": "Carlos", "grado": "8", "materias": ["F√≠sica", "Matem√°ticas"]},
    {"id": 5, "nombre": "Valeria", "grado": "10", "materias": ["Qu√≠mica"]},
]
def mostrar_estudiantes():
    """Muestra todos los estudiantes."""
    print("\n--- Lista de estudiantes ---")
    for e in estudiantes:
        print(f"ID: {e['id']} | Nombre: {e['nombre']} | Grado: {e['grado']} | Materias: {e['materias']}")

def agregar_estudiante():
    """Agrega un nuevo estudiante con los campos solicitados."""
    print("\n--- Agregar Estudiante ---")
    nuevo_id = estudiantes[-1]["id"] + 1 if estudiantes else 1
    nombre = input("Nombre: ")
    grado = input("Grado: ")
    materias = input("Materias favoritas (separadas por comas): ").split(",")

    estudiantes.append({
        "id": nuevo_id,
        "nombre": nombre,
        "grado": grado,
        "materias": [m.strip() for m in materias]
    })

    print("Estudiante agregado correctamente.")

def eliminar_estudiante():
    """Elimina un estudiante seg√∫n su ID."""
    print("\n--- Eliminar Estudiante ---")
    id_eliminar = int(input("ID del estudiante a eliminar: "))

    for e in estudiantes:
        if e["id"] == id_eliminar:
            estudiantes.remove(e)
            print("Estudiante eliminado.")
            return

    print("No existe un estudiante con ese ID.")

def actualizar_estudiante():
    """Actualiza los datos de un estudiante."""
    print("\n--- Actualizar Estudiante ---")
    id_act = int(input("ID del estudiante a actualizar: "))

    for e in estudiantes:
        if e["id"] == id_act:
            nuevo_nombre = input(f"Nuevo nombre ({e['nombre']}): ") or e['nombre']
            nuevo_grado = input(f"Nuevo grado ({e['grado']}): ") or e['grado']
            nuevas_materias = input("Nuevas materias (coma-separadas, vac√≠o para no cambiar): ")

            e["nombre"] = nuevo_nombre
            e["grado"] = nuevo_grado

            if nuevas_materias.strip():
                e["materias"] = [m.strip() for m in nuevas_materias.split(",")]

            print("Datos actualizados.")
            return

    print(" No se encontr√≥ ese ID.")

if __name__ == "__main__":
    while True:
        print("\n=== MEN√ö SISTEMA DE ESTUDIANTES ===")
        print("1. Mostrar estudiantes")
        print("2. Agregar estudiante")
        print("3. Eliminar estudiante")
        print("4. Actualizar estudiante")
        print("5. Salir")

        opcion = input("Elige una opci√≥n: ")

        if opcion == "1":
            mostrar_estudiantes()
        elif opcion == "2":
            agregar_estudiante()
        elif opcion == "3":
            eliminar_estudiante()
        elif opcion == "4":
            actualizar_estudiante()
        elif opcion == "5":
            print("Saliendo del sistema...")
            break
        else:
            print("Opci√≥n inv√°lida. Intenta nuevamente.")



```
